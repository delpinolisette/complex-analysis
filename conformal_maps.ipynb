The goal is to have a notebook for visualizing transformations from the z plane to the w plane. 


```python
"""

visualize confomal(or any) maps from z pl
"""
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
import math

u = linspace(-2, 2, 100)
# print(u)
v = linspace(-2, 2, 100) # evenly spaced numbers on an interval


# making the meshgrid
uu, vv = meshgrid(u, v) # turns a 2 d array of x and y values into plotted grid

# to find the image of a semi circle we would have to have a matrix of values representing the circle. 
# so where x^2 + y^2 = 0 and the imaginary part is strictly positive. 


cu = np.arange(-2,2,.05)
cv = np.arange(-2,2,.05)
print(cu*(cv.T))
circle_x = []
circle_y = []

# keep only the disk equation satisfiers
for u in cu:
    for v in cv:
        if (u**2 + v**2 < 1):
            circle_x.append(u)
            circle_y.append(v)


#plt.scatter(circle_x, circle_y)

circle_x2 = []
circle_y2 = []

# to set the condition that im(z) is greater than 0, set y > 0
for u in cu:
    for v in cv:
        if (u**2 + v**2 < 1) and (v > 0):
            circle_x2.append(u)
            circle_y2.append(v)

plt.scatter(circle_x2, circle_y2)

# multiply by transpose 

z0 = uu + (1j * vv) # delinneate real and complex axes for z plane
#plt.scatter(z0)

#m = 2+3j # first map
z =  ((math.e**z0) - 1)/(z0)# actual transformation 
T = np.arctan2(uu,vv) # argument for each coordinate in complex plane

plt.figure(figsize=(14,6))  # holds both plots
plt.subplot(1,2,1)
#plt.scatter(uu, vv, c=T, s=10, lw = 0)
plt.scatter(uu, vv, c=T)
plt.title('z plane, starting points')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)

plt.subplot(1,2,2)
#plt.scatter(real(z), imag(z),c=T, s=10, lw = 0)
plt.scatter(real(z), imag(z), c=T)
plt.title('w plane, mapped points')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)
plt.gray
```
